```python
import numpy as np

ROW_COUNT = int(input("Enter number of rows:"))
print(ROW_COUNT)

COLUMN_COUNT = int(input("Enter number of columns:"))
print(COLUMN_COUNT)

PIECES_COUNT = int(input("Enter number of pieces:"))
print(PIECES_COUNT)




if ROW_COUNT>COLUMN_COUNT:
    MIN = COLUMN_COUNT
else:
    MIN = ROW_COUNT
    

def create_board():
    board = np.zeros((ROW_COUNT,COLUMN_COUNT))
    return board


def drop_piece(board, row, col, piece):
    board[row][col] = piece

def is_valid_location(board,col):
    return board[ROW_COUNT-1][col] == 0

def get_next_open_row(board,col):
    for r in range(ROW_COUNT):
        if board[r][col] == 0:
            return r

def print_board(board):
    print(np.flip(board,0))
    
def winning_move(board,piece):
    #horizontal
    for c in range(COLUMN_COUNT-(PIECES_COUNT-1)):
        for r in range(ROW_COUNT):
            check = 0
            for p in range(PIECES_COUNT):
                if board[r][c+p]== piece:
                    check+=1
            if check == PIECES_COUNT:
                return True
     
    #vertical
    for c in range(COLUMN_COUNT):
        for r in range(ROW_COUNT-(PIECES_COUNT-1)):
            check = 0
            for p in range(PIECES_COUNT):
                if board[r+p][c]== piece:
                    check+=1
            if check == PIECES_COUNT:
                return True

    #up right diagonal
    for c in range(COLUMN_COUNT-(PIECES_COUNT-1)):
        for r in range(ROW_COUNT-(PIECES_COUNT-1)):
            check = 0
            for p in range(PIECES_COUNT):
                if board[r+p][c+p]== piece:
                    check+=1
            if check == PIECES_COUNT:
                return True

    #up left diagonal
    for c in range(COLUMN_COUNT-(PIECES_COUNT-1)):
        for r in range((PIECES_COUNT-1),ROW_COUNT):
            check = 0
            for p in range(PIECES_COUNT):
                if board[r-p][c+p]== piece:
                    check+=1
            if check == PIECES_COUNT:
                return True

                   
board = create_board()
game_over = False
turn = 0

if ROW_COUNT == 0:
    print("Enter valid rows")
    game_over = True
elif COLUMN_COUNT == 0:
    print("Enter valid columns")
    game_over = True
elif PIECES_COUNT>MIN:
    print("Enter valid Pieces")
    game_over = True



while not game_over:
    print_board(board)
    #Player1 input
    if turn == 0:
        col =int(input("Player1, What column do you want to put your piece?")) 
        if col not in range(0,ROW_COUNT):
            print("Wrong selection, select in the range")
            
        if is_valid_location(board,col):
            row = get_next_open_row(board,col)
            drop_piece(board,row,col,1)
        
            if winning_move(board,1):
                   print("PLAYER 1 Wins!!! Congrats!!!")
                   game_over = True
                   
                   
    #Player2 input
    else:
        col =int(input("Player2, What column do you want to put your piece?")) 
        if col not in range(0,ROW_COUNT):
            print("Wrong selection, select in the range")
    
        if is_valid_location(board,col):
            row = get_next_open_row(board,col)
            drop_piece(board,row,col,2)
            
            if winning_move(board,2):
                   print("PLAYER 2 Wins!!! Congrats!!!")
                   game_over = True
    
    print_board(board)
    
    turn+=1
    turn = turn%2    
        
```
